# Auxiliary output

```python
input_A = keras.layers.Input(shape=[5], name="wide_input")
input_B = keras.layers.Input(shape=[6], name="deep_input")
hidden1 = keras.layers.Dense(30, activation="relu")(input_B)
hidden2 = keras.layers.Dense(30, activation="relu")(hidden1)

concat = keras.layers.concatenate([input_A, hidden2])
output = keras.layers.Dense(1)(concat) # Same as before
aux_output = keras.layers.Dense(1)(hidden2)
model = keras.models.Model(inputs = [input_A, input_B], outputs = [output, aux_output])
```

```python
# Each output will need it's own loss function
model.compile(optimizer= keras.optimizers.SGD(learning_rate=1e-3), 
              loss= [keras.losses.mean_squared_error, keras.losses.mean_squared_error],
              loss_weights=[0.9,0.1]) # Giving more wight to the main output that the auxiliary output

history = model.fit((X_train[:, :5],X_train[:, 2:]),[y_train, y_train], epochs = 20, # Giving each output same labels separately
                   validation_data=((X_valid[:, :5],X_valid[:, 2:]), [y_valid,y_valid]))

total_loss, main_loss, aux_loss = model.evaluate((X_test[:, :5],X_test[:, 2:]), [y_test,y_test]) # model returns separate loss for 

y_pred_main, y_pred_aux = model.predict((X_new[:, :5],X_new[:, 2:])) # Also separate prediction

pd.DataFrame(history.history).plot()
plt.gca().set_ylim(0,1)
plt.grid()
plt.show()
```

# Dynamic neural network

```python
class WideAndDeepModel(keras.models.Model): 
    def __init__(self, units = 30, activation = 'relu', ** kwargs): # This functiion will have the model settings
        super().__init__(** kwargs) # Passes any additional keyword arguments (**kwargs) to the parent for handling.
        self.hidden1 = keras.layers.Dense(units, activation=activation)
        self.hidden2 = keras.layers.Dense(units, activation=activation)
        self.main_output = keras.layers.Dense(1)
        self.aux_output = keras.layers.Dense(1)
    def call(self, inputs): # This will hold the model structures
        input_A, input_B = inputs
        hidden1 = self.hidden1(input_B)
        hidden2 = self.hidden2(hidden1)
        concat = keras.layers.concatenate([input_A, hidden2])
        main_output = self.main_output(concat)
        aux_output = self.aux_output(hidden2)
        return main_output, aux_output
```

```python
model = WideAndDeepModel()
```

# Using callback to save checkpoints in the case of training large dataset

```python
keras.backend.clear_session()
# Models: All defined model architectures and their associated variables are removed from memory.
# Variables: Any standalone variables you created using TensorFlow operations (e.g., tf.Variable) are also cleared.
# Optimizer State: If you were training a model, the optimizer's state (which tracks progress during training) is also reset.

np.random.seed(42)
tf.random.set_seed(42)

model = keras.models.Sequential([
    keras.layers.Dense(30, activation="relu", input_shape=[8]),
    keras.layers.Dense(30, activation="relu"),
    keras.layers.Dense(1)
])    
```

# ModelCheckpoint callback

```python
model.compile(loss="mse", optimizer=keras.optimizers.SGD(learning_rate=1e-3))
checkpoint_cb = keras.callbacks.ModelCheckpoint("saving_best_parameter_during_training", save_best_only=True)#The Model
                                                    #Checkpoint callback saves checkpoints of your model at regular intervals during
                                                    #training, by default at the end of each epoch
history = model.fit(X_train_scaled, y_train, epochs=10,
                    validation_data=(X_valid_scaled, y_valid),
                    callbacks=[checkpoint_cb])
#model = keras.models.load_model("saving_best_parameter_during_training") # Getting the checkpoint
#mse = model.evaluate(X_test, y_test)
```

```python
# Save_best_only = True will only save the model if it does good on validation set
# So it will save the model before it start overfitting the data
```

# Early stopping callback

```python
model.compile(loss="mse", optimizer=keras.optimizers.SGD(learning_rate=1e-3))

early_stopping_cb = keras.callbacks.EarlyStopping(patience=10, restore_best_weights=True)
# If restore_best_weights=True, the callback:
    # Keeps track of the model's weights that produced the best validation performance so far.
    # Upon stopping, automatically reverts the model's weights back to those best weights.
history = model.fit(X_train_scaled, y_train, epochs=20,
                   validation_data=(X_valid_scaled, y_valid),
                   callbacks = [checkpoint_cb, early_stopping_cb])
mse_test = model.evaluate(X_test_scaled, y_test)
```

# Custom callback

```python
# A callback that measures the overfitting proportion
# If the ratio of validation loss to training loss increases
#significantly (e.g., well above 1.0), it's a strong signal of potential overfitting

class OverFittingRatio(keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs):
        print("/nVal/Train:{:.2f}".format(logs['val_loss'] / logs['loss']))
```

```python
val_train_ratio_cb = OverFittingRatio()
history = model.fit(X_train_scaled, y_train, epochs=1,
                    validation_data=(X_valid_scaled, y_valid),
                    callbacks=[val_train_ratio_cb])
```

```python
#The on_epoch_end method in Keras callbacks is specifically designed to execute code at the end of each training epoch
# As you might expect, you can implement on_train_begin(), on_train_end(),
# on_epoch_begin(), on_epoch_begin(), on_batch_end() and on_batch_end().
# Moreover, callbacks can also be used during evaluation and predictions, should you
# ever need them (e.g., for debugging). In this case, you should implement
# on_test_begin(), on_test_end(), on_test_batch_begin(), or
# on_test_batch_end() (called by evaluate()), or on_predict_begin(), on_pre
# dict_end(), on_predict_batch_begin(), or on_predict_batch_end() (called by
# predict()).
```

# Hyperparameter tuning

```python
# In order to tune hyperparameter we need to make the DNN work as scikit-learn model
# Author used keras_reg = keras.wrappers.scikit_learn.KerasRegressor(build_model) but some reason I do not have it.
# I downloaded a scikeras library which act as the same way

from scikeras.wrappers import KerasRegressor

def build_model(n_hidden = 1, n_neurons = 30,learning_rate=3e-3, input_shape=[8]):
    model = keras.models.Sequential()
    model.add(keras.layers.InputLayer(input_shape=input_shape))
    for i in range(n_hidden):
        model.add(keras.layers.Dense(n_neurons, activation='relu'))
    model.add(keras.layers.Dense(1))
    model.compile(loss=keras.losses.mean_squared_error,optimizer=keras.optimizers.SGD(learning_rate=learning_rate))
    return model

keras_reg = KerasRegressor(build_model, n_neurons=90, n_hidden=3, learning_rate=0.0013274989938894624) # It will make the keras model act as scikit-learn model. 
                                        # As a result we can use score, predict ect. the same thing in keras model too

keras_reg.fit(X_train_scaled,y_train,epochs=30,
         validation_data=(X_valid_scaled, y_valid),
         callbacks=[keras.callbacks.ModelCheckpoint('keras_reg_checkpoint', save_best_only=True)])
mse = keras_reg.score(X_test_scaled, y_test) # It does not have evaluate so have to use score
y_pred = keras_reg.predict(X_new)
```
