# First downloading the latest dataset on house_price from github

```python
import os
import tarfile
import urllib.request

DOWNLOAD_ROOT = "https://raw.githubusercontent.com/ageron/handson-ml2/master/"
HOUSING_PATH = os.path.join("datasets", "housing")
HOUSING_URL = DOWNLOAD_ROOT + "datasets/housing/housing.tgz"

def fetch_housing_data(housing_url=HOUSING_URL, housing_path=HOUSING_PATH):
    if not os.path.isdir(housing_path):
        os.makedirs(housing_path)
    tgz_path = os.path.join(housing_path, "housing.tgz")
    urllib.request.urlretrieve(housing_url, tgz_path)#Retrives data from the github's writter's file
    housing_tgz = tarfile.open(tgz_path)
    housing_tgz.extractall(path=housing_path)
    housing_tgz.close()       
```

# Creating custom train test sets

```python
def split_train_test_set(data, test_ratio):
    suffled_indices = np.random.permutation(len(data))
    test_set_size = int(len(data) * test_ratio)
    test_indices = suffled_indices[:test_set_size]
    train_indices = suffled_indices[test_set_size:]

    return data.iloc[train_indices], data.iloc[test_indices]
```

## Making sure that test set in unique by giving them unique hash(like finger print)

```python
from zlib import crc32

def check_test_set(indentifier, test_ratio):
    return crc32(np.int64(indentifier)) & 0xffffffff < test_ratio * 2*32

def split_train_test_by_id(data, id_column, test_ratio): # Id column will be the index number
    ids = data[id_column]
    in_test_set = ids.apply(lambda id_: check_test_set(id_, test_ratio)) # Gives a bool list
    return data.loc[~in_test_set], data.loc[in_test_set] # ( ~ ) makes the in_test_set to false, meaning it is the train set
```

# Custom estimator

```python
# We will make a estomator to add extra attribes(combines atrribs one)
# As we will only add attribs and nothing to learn the fit() part will be blank
from sklearn.base import BaseEstimator,BiclusterMixin

class CombinedAttributesAdder(BaseEstimator, BiclusterMixin):
    def __init__(self, add_bedrooms_per_room = True): # To filter the output this attribs was added, according to our target we can add as many as we needed
        self.add_bedrooms_per_room = add_bedrooms_per_room
    def fit(self, X, y = None):
        return self # We do not want learn anything
    def transform(self, X, y=None):
        rooms_per_households = X[:, rooms_ix]/ X[:, households_ix] # _ix represents the indices, which is define in the next cell
        population_per_households = X[:, population_ix] / X[:, households_ix]
        if self.add_bedrooms_per_room:
            bedrooms_per_households = X[:, bedrooms_ix] / X[:, households_ix]
            return np.c_[X, rooms_per_households, population_per_households, bedrooms_per_households]
        else:
            return np.c_[X, rooms_per_households, population_per_households]
```

```python
rooms_ix, bedrooms_ix, population_ix, households_ix = 3,4,5,6
attr_adder = CombinedAttributesAdder(add_bedrooms_per_room = False)# We will not add bedrooms as it do not corr much with the target value
housing_extra_attribs = attr_adder.transform(housing.values) # This will add the attribs but in numpy form and will not have any labels
```

# Standardization or Normalization?

```python
from sklearn.preprocessing import StandardScaler # Standardize 
from sklearn.preprocessing import MinMaxScaler # Normalization
"""One of the most important transformations you need to apply to your data is feature
scaling. With few exceptions, Machine Learning algorithms don’t perform well when
the input numerical attributes have very different scales. This is the case for the hous‐
ing data: the total number of rooms ranges from about 6 to 39,320, while the median
incomes only range from 0 to 15. Note that scaling the target values is generally not
required."""

"""Normalization:
    
    Description:scales data in range of 0 to 1
    Power: Makes the scaling even as a result the model performance improves
    
    Standardization:
    
    Description:makes the mean 0 and the standard deviation to 1
    Power: std is 1, so it is less affected by outliers compared to normalization"""
```

# Pipelines

```python
from sklearn.pipeline import Pipeline

num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median')), # Estimator
    ('attr_adder', CombinedAttributesAdder(add_bedrooms_per_room=False)), # Estimator
    ('scale', StandardScaler()), # Transformer
])
# Everyone but the last have to be estimator and the last one is the transformer, that has only trasnform() attribute, 
# the rest has fit_transform() attribute
housing_num_tr = num_pipeline.fit_transform(housing_num)
```

```python
from sklearn.compose import ColumnTransformer

# Unlike pipeline column transformer combine both categorical and numerical attribute transformation.
num_attribs = list(housing_num)
cat_attribs = ['ocean_proximity']
full_pipeline = ColumnTransformer([
    ('num', num_pipeline, num_attribs),
    ('cat', OneHotEncoder(), cat_attribs),
])

housing_prepared = full_pipeline.fit_transform(housing)
```

# Implementing pipeline on linear_reg and decision_tree

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(housing_prepared, housing_labels)
```

```python
from sklearn.metrics import mean_squared_error

y_pred_lin = lin_reg.predict(housing_prepared)
scores = mean_squared_error(housing_labels, y_pred_lin)
lin_rmse = np.sqrt(scores) # Measuring the error of linrear_reg
lin_rmse
```

```python
from sklearn.tree import DecisionTreeRegressor

tree_reg = DecisionTreeRegressor(max_depth=3, random_state=42)
tree_reg.fit(housing_prepared, housing_labels)
y_pred_tree = tree_reg.predict(housing_prepared)
score = mean_squared_error(housing_labels, y_pred_tree)
tree_rmse = np.sqrt(score)
tree_rmse
```

# Using cross validation to measure best model

```python
from sklearn.model_selection import cross_val_score

lin_neg_scores = cross_val_score(lin_reg, housing_prepared, housing_labels,
                                   scoring = 'neg_mean_squared_error',
                                   cv = 3)
lin_rmse_scores = np.sqrt(-lin_neg_scores) # Scores are measured in neg so we have to give a neg sign to sqrt it
lin_rmse_scores
```

```python
tree_neg_scores = cross_val_score(tree_reg, housing_prepared, housing_labels,
                                  scoring = 'neg_mean_squared_error',
                                  cv = 3)
tree_rmse_scores = np.sqrt(-tree_neg_scores)
tree_rmse_scores
```

```python
def display_scores(scores):
    print("Scores:", scores)
    print("Mean:", scores.mean())# Mean is for seeing what the model overall performance is
    print("Std:", scores.std())# To see how well each scores are
```

```python
display_scores(tree_rmse_scores)
```

```python
display_scores(lin_rmse_scores)
```

# Finding hyperparameter

```python
from sklearn.model_selection import GridSearchCV

# In order to implement the grid_search we will use random forest, also can implement it on decision tree
```

```python
from sklearn.ensemble import RandomForestRegressor

forest_reg = RandomForestRegressor()
param_grid = [
    {'n_estimators' : [3,10,30], 'max_features' : [3,4,5]},
    {'bootstrap' : [False], 'n_estimators' : [3,10,30], 'max_features': [3,4,5,6]}
]

grid_search = GridSearchCV(forest_reg, param_grid, cv=5,
                           scoring='neg_mean_squared_error',
                           return_train_score=True,
)

grid_search.fit(housing_prepared, housing_labels)
```

```python
grid_search.best_params_ # Shows the best patameters that scored well on the dataset
```

```python
grid_search.best_estimator_ # The best estimator with best params
```

```python
# Seeing what each estimators scores was 

cvres = grid_search.cv_results_
for mean_score, params in zip(cvres['mean_test_score'], cvres['params']):
    print(np.sqrt(-mean_score), params)
```
